```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
import numpy as np
```

```python
df = pd.read_csv("/content/AMD_cleaned.csv", parse_dates=["Date"], index_col="Date")
```

```python
df_period = df.loc["2020-01-01":"2024-12-31"].copy()
```

```python
df_period["LogReturn"] = np.log(df_period["Close"] / df_period["Close"].shift(1))
```

```python
df_period = df_period.dropna()
```

```python
plt.figure(figsize=(10,4))
plot_acf(df_period["Close"], lags=50)
# plt.title("Autocorrelation of AMD Closing Prices (2020–2024)")
plt.tight_layout()
plt.show()

# ACF for Log Returns
plt.figure(figsize=(10,4))
plot_acf(df_period["LogReturn"], lags=50)
# plt.title("Autocorrelation of AMD Log Returns (2020–2024)")
plt.tight_layout()
plt.show()
```
